## Create the best plot using polynomials!

Here is a plot with a clearly non-linear shape. Begin to think about what degree polynomial you believe will fit it best.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def pol(x):
    return x * np.cos(x)

x = np.linspace(0, 12, 100)
rng = np.random.RandomState(1234)
rng.shuffle(x)
x = np.sort(x[:25])
y = pol(x) + np.random.randn(25)*2

plt.scatter(x, y, color='green', s=50, marker='.')

plt.show();
```

Now it's time to determine the optimal degree of polynomial features for a model that is fit to this data     
- Plot a polynomial function using `PolynomialFeatures` for polynomials up until the second, third and fourth degrees. Make sure to label each degree polynomial in your plot.   
- Print out the $R^2$ value for each of the three results, then decide which degree is best.

## Solution

```python
import pandas as pd
x_values = pd.DataFrame(x)
y_values = pd.DataFrame(y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
fig = plt.figure(figsize=(10,6))
plt.scatter(x_values, y_values)

reg = LinearRegression()
reg.fit(x_values, y_values)
r2 = r2_score(y_values, reg.predict(x_values))
print(r2)
plt.plot(x_values, reg.predict(x_values), label = f"Polynomial: {i}, R-squared: {r2}")
plt.show()
```

```python


for i in range(2, 5):
    poly = PolynomialFeatures(i)
    X_fin = poly.fit_transform(x_values)
#     print(X_fin)
    reg_poly = LinearRegression().fit(X_fin, y_values)
    predict_X = reg_poly.predict(X_fin)
    y_poly_pred = reg_poly.predict(X_fin)
    r2 = r2_score(y_values, y_poly_pred)
    print(r2)
    fig = plt.figure(figsize=(10,6))
    plt.scatter(x_values, y_values)
    plt.plot(x_values, y_poly_pred, label = f"Polynomial: {i}, R-squared: {r2}")
    plt.legend()
    plt.show()
```
